## Plotting basics
- mathplotlib
- seaborn
- pandas plot() method

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/titanic.csv")
df.head()
```

```python
df.info()
```

## matlab - syntax in matplotlib

```python
# varje kolumn är en 'serie', vi kan plocka ut 'age' - för att sedan plotta den.
import matplotlib.pyplot as plt

plt.hist(df["age"])
```

```python
import matplotlib.pyplot as plt

plt.hist(df["age"], bins = 50)
plt.xlable("age")
plt.ylable("count")
plt.title("Histogram of ages in titanic dataset")
```

## OOP approach in matplotlib

```python
# fig innehåller flera ax objekt, (hur många grafer, bestäms inom parantesen)

fig, ax  = plt.subplots(1)
ax.hist(df["age"], bins = 50);
ax.set(title = "Histogram of ages in titanic dataset", xlabel ="age", ylabel="count")
```

## bar chart in matplotlib

```python
df.columns
```

```python
classes = df["class"].value_counts().reset_index()
classes
```

```python
fig, ax = plt.subplots(1)
ax.bar(x = classes["class"], height = classes ["count"])
```

```python
cohorts = ["DE23", "DE24", "AI23", "JAVA24"]
numbers_students = [29, 38, 28, 20]

fig, ax = plt.subplots(1)
ax.bar( x = cohorts, height = numbers_students)
```

## bar chart in seaborn

```python
import seaborn as sns

classes
```

```python
#matplotlib
# fig, ax = plt.subplots(1)
# ax.bar(x = classes["class"], height = classes ["count"])
# ax.set(xlabel = "class", ylabel = "count", title = "classes onboard of titanic")

# seaborn
ax = sns.barplot(data = classes, x = "class", y = "count" )
ax.set(title = "classes onboard in titanic")
```

```python
# multiple - för att få staplarna separerade, kde - för att få en kurva över
sns.histplot(df, x = "age", hue ="sex", multiple = "dodge", kde = True)
```
